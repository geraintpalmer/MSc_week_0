```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style('whitegrid')
sns.set_palette("PuOr")
import math
import random
import pandas as pd
```

Coordinates on a plane

```python
fig, ax = plt.subplots(1, figsize=(8, 6))
xmin = -0.05
xmax = 2
ymin = -0.05
ymax = 1.5
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
ax.plot([0, 0], [ymin, ymax], c='black')
ax.plot([xmin, xmax], [0, 0], c='black')
ax.set_xlabel('x', fontsize=18)
ax.set_ylabel('y', rotation=0, fontsize=18)
ax.scatter([1.3], [0.75], s=100)
sns.despine()
plt.tight_layout()
fig.savefig('coordinate_system.pdf')
```

Quadratic

```python
fig, ax = plt.subplots(1, figsize=(8, 6))
xmin = -2
xmax = 3
ymin = -3
ymax = 15
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
ax.plot([0, 0], [ymin, ymax], c='black')
ax.plot([xmin, xmax], [0, 0], c='black')
ax.set_xlabel('x', fontsize=18)
ax.set_ylabel('y', rotation=0, fontsize=18)
xs = np.linspace(xmin, xmax, 200)
ys = [3*x**2 - 1.5*x -2 for x in xs]
ax.plot(xs, ys, linewidth=3)
ax.set_title(r'$y = 3x^2 - \frac{3}{2}x - 2$', fontsize=24)
sns.despine()
plt.tight_layout()
fig.savefig('quadratic.pdf')
plt.show()
```

Tangent curves

```python
fig, ax = plt.subplots(1, figsize=(8, 6))
xmin = -0.05
xmax = 3.05
ymin = -2.05
ymax = 4.05
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
ax.plot([0, 0], [ymin, ymax], c='black')
ax.plot([xmin, xmax], [0, 0], c='black')
ax.set_xlabel('x', fontsize=18)
ax.set_ylabel('y', rotation=0, fontsize=18)
xs = np.linspace(xmin, xmax, 200)
ys = [x**2 - x for x in xs]
ax.scatter([1], [0], s=100)
ax.scatter([2], [2], c='darkgreen', s=100)
ax.plot(xs, ys, linewidth=3)
ax.plot([xmin, xmax], [2*xmin - 2, 2*xmax - 2], linewidth=3, c='darkgreen')
sns.despine()
plt.tight_layout()
fig.savefig('tangent_line.pdf')
```

Exponential

```python
fig, ax = plt.subplots(1, figsize=(8, 6))
xmin = -1.55
xmax = 3.05
ymin = -0.05
ymax = 21
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
ax.plot([0, 0], [ymin, ymax], c='black')
ax.plot([xmin, xmax], [0, 0], c='black')
ax.set_xlabel('x', fontsize=18)
ax.set_ylabel('y', rotation=0, fontsize=18)
xs = np.linspace(xmin, xmax, 200)
ys = [math.exp(x) for x in xs]
ax.plot(xs, ys, linewidth=3)
ax.set_title(r'$y = e^x$', fontsize=24)
sns.despine()
plt.tight_layout()
fig.savefig('exp.pdf')
```

Logs in stats

```python
countries = pd.read_excel('countries of the world.xls')
```

```python
countries['Log Area'] = countries['Area'].apply(lambda x: math.log(x, 10))
countries['Log Population'] = countries['Population'].apply(lambda x: math.log(x, 10))
```

```python
fig, ax = plt.subplots(1, figsize=(6, 6))
countries.plot(kind='scatter', x='Area', y='Population', ax=ax, alpha=0.7, edgecolor='black', s=50)
ax.set_xlabel('Area', fontsize=18)
ax.set_ylabel('Population', fontsize=18)
sns.despine()
fig.savefig('beforelog.pdf')
```

```python
fig, ax = plt.subplots(1, figsize=(6, 6))
countries.plot(kind='scatter', x='Log Area', y='Log Population', ax=ax, alpha=0.7, edgecolor='black', s=50)
ax.set_xlabel(r'$\log_{10}{(Area)}$', fontsize=18)
ax.set_ylabel(r'$\log_{10}{(Population)}$', fontsize=18)
sns.despine()
fig.savefig('afterlog.pdf')
```
